```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('maindata/01_제공데이터/2020 빅콘테스트 데이터분석분야-챔피언리그_2019년 실적데이터_v1_200818.xlsx', skiprows=[0])
df = df.rename(columns = {"Unnamed: 1":'time',"Unnamed: 2":"ampm"})
df
```

```python
df1 = df.copy()
```

```python
df1.info()
```

```python
df1.describe(include = "all")
```

```python
df1[df1["상품군"] != '무형']
```

```python
idx_무형 = df1[df1["상품군"]== "무형"].index
data = df1.drop(idx_무형)
data
```

```python
data.info()
```

```python
t = data[data['노출(분)'] > 0]
t1 = t[["방송일시", 'time',"ampm",'노출(분)', '마더코드','판매단가','취급액']]
t1
```

```python
data2 = pd.merge(data, t1,how = "left", left_on = ('방송일시','time',"ampm"), right_on = ('방송일시','time',"ampm"))
data2
```

```python
data2.isnull().sum()
```

```python
data2.columns
```

```python
data3 = data2[['방송일시', 'time', 'ampm', '노출(분)_y', '마더코드_x', '상품코드', '상품명', '상품군','판매단가_x', '취급액_x']]
data3
```

```python
d = [ i[:3] for i in data3['time'].tolist()]
d
```

```python
data3['hour'] = pd.to_datetime(data3["time"], format = "%H:%M:%S").dt.hour
data3
```

```python
data3.columns
```

```python
data3 = data3[['방송일시', 'time', 'hour', 'ampm', '노출(분)_y', '마더코드_x', '상품코드', '상품명', '상품군',
       '판매단가_x', '취급액_x' ]]
```

```python
data4 = data3[['방송일시','hour', 'ampm', '노출(분)_y', '마더코드_x', '상품코드', '상품명', '상품군',
       '판매단가_x', '취급액_x' ]]
```

```python
data3_group = data4.groupby(['방송일시','hour', 'ampm','마더코드_x', '상품코드', '상품명', '상품군','판매단가_x'])['노출(분)_y','취급액_x'].sum()
data3_group = data3_group.reset_index()
```

```python
data3_group.to_csv("test_g.csv")
```

```python
time = data3.drop_duplicates(subset = ['방송일시','hour','ampm','마더코드_x','상품코드','상품명'], keep ="first")
time
```

```python
data3_g_time = pd.merge(data3_group, time, how="inner", on=('방송일시','hour','ampm','마더코드_x','상품코드','상품명'))
data3_g_time
```

```python
data3_g_time.columns
```

```python
data3_g_time = data3_g_time[['방송일시','time', 'hour', 'ampm', '노출(분)_y_x','마더코드_x', '상품코드', '상품명', '상품군_x', '판매단가_x_x',
        '취급액_x_x']].copy()
```

```python
data3_g = data3_group.copy()
```

```python
data3_g_time = data3_g_time.rename(columns = {'노출(분)_y_x':'노출시간','마더코드_x':"마더코드", '상품군_x':'상품군', '판매단가_x_x':"판매단가",
        '취급액_x_x':'취급액'})
data3_g_time
```

```python
data3_g = data3_g.rename(columns = {'노출(분)_y':'노출시간','마더코드_x':"마더코드", '상품군_x':'상품군', '판매단가_x':"판매단가",
        '취급액_x':'취급액'})
```

```python
import numpy as np
data3_g["판매개수"] = np.ceil(data3_g["취급액"]/data3_g['판매단가'])
data3_g
```

```python
data3_g_time["판매개수"] = np.ceil(data3_g_time["취급액"]/data3_g['판매단가'])
data3_g_time
```

```python
for i in range(len(data3_g)):
    if data3_g['ampm'][i] == "PM":        
        if data3_g["hour"][i] != 12:
            data3_g["hour"][i] += 12
    elif data3_g['ampm'][i] == "AM":
        if data3_g["hour"][i] == 12:
            data3_g["hour"][i] = 0
            
data3_g
```

```python
for i in range(len(data3_g_time)):
    if data3_g_time['ampm'][i] == "PM":        
        if data3_g_time["hour"][i] != 12:
            data3_g_time["hour"][i] += 12
    elif data3_g_time['ampm'][i] == "AM":
        if data3_g_time["hour"][i] == 12:
            data3_g_time["hour"][i] = 0
            
data3_g_time
```
