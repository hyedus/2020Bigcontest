```python
# 제공데이터 마더코드mcode 상품코드item

mcode = df["마더코드"].drop_duplicates().tolist()
item = df["상품코드"].drop_duplicates().tolist()

print(len(mcode))
print(len(item))
```

```python
# 평가데이터 마더코드 mcode_t, 상품코드 item_t

mcode_t = df2['마더코드'].drop_duplicates().tolist()
item_t = df2["상품코드"].drop_duplicates().tolist()

print(len(mcode_t))
print(len(item_t))
```

```python
# 마더코드 비교 (평가데이터에만 있는 코드)

mlist = []
for i in mcode_t:
    if i not in mcode:
        mlist.append(i)
        
print(len(mlist))
print(mlist)
```
