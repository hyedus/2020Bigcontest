```python
text = rtr["context 2"]
from konlpy.tag import Kkma
kkma = Kkma()

k_tags = []
for c in text:
    kkma_tokens = kkma.morphs(c)
    k_tags.append(kkma_tokens)
    
    
```

```python

kkmaTag = []
for tags in k_tags:
    taglist = []
    for t in tags:
        taglist += kkma.pos(t)
    kkmaTag.append(taglist)
    

#     for tag in kkmaTag:
#         wordlist = []
#         if tag[1] in ["NNG","NNP"]:
#                if len(tag[0]) > 1:
kkmaTag
```

```python
word = []
for tags in kkmaTag:
    wordlist = []
    for tag in tags:        
        if tag[1] in ["NNG"]:
            if len(tag[0]) > 1:
                wordlist.append(tag[0])
    word.append(wordlist)
    
word
```

```python
words = sum(word, [])
```
