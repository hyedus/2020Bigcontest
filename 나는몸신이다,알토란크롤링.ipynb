## 나는 몸신이다, 크롤링저장

```python
import requests
from bs4 import BeautifulSoup
result_list = []

for idx in id_list:
    num = idx["id"]
    page_url = "http://www.ichannela.com/program/template/video_detailPage.do"
    params = {"publishId" : num , "subCateCode": "050014"}
    resp = requests.get(page_url, params=params)
    soup = BeautifulSoup(resp.text, "html.parser")
    
    article = soup.find("div", class_="article_detail").text
    date = soup.find("span", style ="margin-right: 5px;").text
    
    result_list.append({
        "날짜": date,
        '내용': article})
    
import pandas as pd    
result = pd.DataFrame(result_list)
result
```

```python
import pandas as pd
df = pd.read_csv("data/알토란주소들.csv", encoding="cp949")
page_url = df["주소"].tolist()
```

```python
from selenium import webdriver

from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.alert import Alert
import time 
rtr = []
driver = webdriver.Chrome()
for idx, url in enumerate(page_url):
    driver.get(url)
    time.sleep(5)
    date = df['날짜'][idx]
    #tbody_b = driver.find_element_by_tag_name("tbody")
    text = driver.find_element_by_tag_name("table").text
    #contxt = text[2].text
    #text = trs[2].text
    rtr.append({'date': date,
               "context": text})
    time.sleep(2)
    print(url, "완료")


rtrdf = pd.DataFrame(rtr)
rtrdf.to_excel("data/rtrdf.xlsx", encoding="utf-8")
```
