```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('maindata/01_제공데이터/2020 빅콘테스트 데이터분석분야-챔피언리그_2019년 실적데이터_v1_200818.xlsx', skiprows=[0])
df = df.rename(columns = {"Unnamed: 1":'time',"Unnamed: 2":"ampm"})
df
```

```python
df1 = df.copy()
```

```python
for i in range(len(df1)):
    if (df1["노출(분)"][i] != np.nan):
        continue
    elif (df1["노출(분)"][i] == np.nan) and (df1["마더코드"][i] == df1["마더코드"][i-1]):
        df1.fillna(method='ffill',inplace=True)
    elif (df1["노출(분)"][i] == np.nan) and (df1["상품코드"][i] == df1["상품코드"][i-1]):
        df1.fillna(method='ffill',inplace=True)
    elif (df1["노출(분)"][i] == np.nan) and (df1["상품명"][i] == df1["상품명"][i-1]):
        df1.fillna(method='ffill',inplace=True)
    else:
        continue
```

```python
df1.info()
```

```python
t = df1[df1['노출(분)'] >0]
t1 = t[["방송일시", 'time','노출(분)', '마더코드']]
t1
```

```python
data1 = pd.merge(df1, t1,how = "left", left_on = ('방송일시','time','마더코드'), right_on = ('방송일시','time','마더코드'))
```

```python
data1.to_csv("check.csv")
```
